# Python kernel check

Run all cells to confirm your `single_cell` conda kernel is working. See `student-setup.md` Steps 12-13.

```python
import numpy as np
import pandas as pd
import anndata as ad
import matplotlib.pyplot as plt

print("numpy:", np.__version__)
print("pandas:", pd.__version__)
print("anndata:", ad.__version__)
```

```python
# toy AnnData object: 5 cells x 4 genes
X = np.random.default_rng(0).poisson(5, size=(5, 4)).astype(float)
adata = ad.AnnData(X=X)
adata.obs_names = [f"cell_{i}" for i in range(adata.n_obs)]
adata.var_names = [f"gene_{i}" for i in range(adata.n_vars)]
adata
```

```python
# quick plot to confirm matplotlib renders in this kernel
plt.figure(figsize=(4, 3))
plt.imshow(adata.X, cmap="viridis", aspect="auto")
plt.xlabel("genes")
plt.ylabel("cells")
plt.title("toy counts")
plt.colorbar()
plt.show()
```
